```python
import pandas as pd

#데이터 불러오기
df = pd.read_csv("smilestyle_dataset.tsv", sep="\t")
# display(df.head())

display(df)

#각 열의 결측치(NaN) 비율을 계산하여 표시
display(df.isna().mean())
display(df.describe())
print(df.shape)

##초딩이랑 중딩이 결측치 비율이 제일 적어서 얘네 둘을 사용하자
```

```python
print(type(choding_column))
print(len(choding_column))

pairs = []


# print(choding_column.isnull())
# count = 0 
# for i in choding_column.isnull():
#     if i == True:
#         count += 1
# print(count)

null_list = choding_column.isnull()

for i in range(len(choding_column) -1):
    if (null_list[i] == True) or (null_list[i+1] == True):
        continue
    pair = [choding_column[i], choding_column[i+1]]
    pairs.append(pair)
# print(count)
print(pairs)

```

```python
import json

json_data = json.dumps(pairs, ensure_ascii=False)

print(json_data)
```

```python
# JSON 파일로 저장
with open('choding_data.json', 'w') as json_file:
    json_file.write(json_data)
```

```python
import pandas as pd
import json

def dump_dataset(name):
    df = pd.read_csv("smilestyle_dataset.tsv", sep="\t")

    display(df.head())
    
    dataset_column = df[name]

    null_list = dataset_column.isnull()

    pairs = []

    for i in range(len(dataset_column) -1):
        if (null_list[i] == True) or (null_list[i+1] == True):
            continue
        pair = [dataset_column[i], dataset_column[i+1]]
        pairs.append(pair)
        

    json_data = json.dumps(pairs, ensure_ascii=False)

    print(json_data)

    # JSON 파일로 저장
    with open(f'{name}.json', 'w') as json_file:
        json_file.write(json_data)

dump_dataset('joongding')
```
